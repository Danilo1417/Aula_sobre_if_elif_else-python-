Atividade 01

```python
#atividade 01
numero = float(input("Digite um número:"))
if numero > 0:
  print("Número positivo")
elif numero == 0:
  print("Número nulo")
else:
  print("Número negativo")
```

Atividade 02

```python
#Ativade 02
numero = int(input("Digite um número:"))
if numero % 2 == 0:
  print("Número é par")
else:
  print("Número é ímpar")
```

Atividade 03

```python
#Atividade 03
nota = float(input("Digite a nota do aluno: "))
if nota >= 7:
  print("Aprovado")
elif nota >= 5:
  print("Recuperação")
else:
  print("Reprovado")
```
